```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels = pd.read_csv('dog-breed-identification\labels.csv')
```

```python
labels.groupby('breed').count().hist()
plt.title('Number of samples per class')
plt.xlabel('Number of train samples')
plt.ylabel('Number of class')
plt.show()
```

```python
print('mean:')
print(labels.groupby('breed').count().mean())
print('max:')
print(labels.groupby('breed').count().max())
print('min:')
print(labels.groupby('breed').count().min())
```

```python
INPUT_SIZE = 224
NUM_CLASSES = 120
```

```python
selected_breed_list = list(labels.groupby('breed').count().sort_values(by='id', ascending=False).head(NUM_CLASSES).index)
labels = labels[labels['breed'].isin(selected_breed_list)]
```

```python
print ("Number of classes: %d" % len(set(labels.breed)))
print ("Missing labels: " + str(labels.isnull().values.any()))
```

```python
class_freq = labels.breed.value_counts()
```

```python
yy = pd.value_counts(labels['breed'])

fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
sns.set_style("whitegrid")

ax = sns.barplot(x = yy.index, y = yy, data = labels)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, fontsize = 8)
ax.set(xlabel='Dog Breed', ylabel='Count')
ax.set_title('Distribution of Dog breeds')
```
